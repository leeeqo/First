# Постановка задачи
Используя схему бегущего счета и итерационные методы решить задачу:
\begin{equation*}
\begin{cases}
   \displaystyle \frac{\partial u}{\partial t} - \frac{e^u}{1+e^{2u}} \frac{\partial u}{\partial x} = 0, \; \; -1 \le x < 0 \\ 
   u(x,0) = - \sin\pi x, \\
   u(0,t) = e^{-t} - 1
\end{cases}
\end{equation*}


# Аналитическое решение задачи
Предположим, что $∃V(x,t,u)=0$ – решение уравнения в неявном виде, где V – дифференцируемая функция своих переменных, причем $\frac{\partial V}{\partial u} \neq 0$. Тогда верно:
$$
\displaystyle \frac{\partial u}{\partial t} = - \frac{\frac{\partial V}{\partial t}}{\frac{\partial V}{\partial u}}, \; \; \frac{\partial u}{\partial x} = - \frac{\frac{\partial V}{\partial x}}{\frac{\partial V}{\partial u}}
$$
и квазилинейное уравнение вида 
$$
\displaystyle a(x,t,u) \frac{\partial u}{\partial t} + b(x,t,u) \frac{\partial u}{\partial x} = f(x,t,u)
$$
записывается в виде линейного однородного уравнения для $V(x,t,u)$:
$$
\displaystyle a(x,t,u) \frac{\partial V}{\partial t} + b(x,t,u) \frac{\partial V}{\partial x} + f(x,t,u) \frac{\partial V}{\partial u} = 0
$$
Введем понятие характеристик как линий в пространстве $(x,t,u)$, касательные к которым в каждой точке параллельны вектору 
$$
\vec A(x,t,u)= \{ a(x,t,u),b(x,t,u),f(x,t,u) \}
$$ 
в той же точке:
$$
\displaystyle \{ t=t(\tau), \; x=x(\tau), \; u=u(\tau) \} \Leftrightarrow \{ \frac{dt}{d\tau}=a, \; \frac{dx}{d\tau}=b, \; \frac{du}{d\tau}=f, \;  \}
$$
Тогда характеристическое уравнение примет вид:
$$
\displaystyle \frac{dt}{a(x,t,u)}=\frac{dx}{b(x,t,u)}=\frac{du}{f(x,t,u)}
$$
В данной задаче $a(x,t,u)=1$, $b(x,t,u)=-\frac{e^u}{1+e^{2u}}$, $f(x,t,u)=0$ и характеристическое уравнение принимает вид:
$$
\displaystyle \frac{dt}{1}=- \frac{dx}{\frac{e^u}{1+e^{2u}}}=\frac{du}{0}
$$
Будем искать решение поставленной задачи в неявном виде, т.е. в виде $V = F(\Psi_1(x,t,u), \; \Psi_2(x,t,u))$ = 0. Найдем первые интегралы:
$$
\begin{cases}
   \Psi_1(x,t,u)=C_1 \\
   \Psi_2(x,t,u)=C_2
\end{cases}
$$
$$
\begin{cases}
   u = C_1 \\
   x + t \frac{1+e^{2u}}{e^u} = C_2
\end{cases}
$$
где $C_1, C_2$ - произвольные константы.

Тогда решение исходного уравнения в неявном виде принимает следующий вид:
$$
V = F\left( u, x + t \frac{1+e^{2u}}{e^u} \right) = 0
$$

# Построение характеристик
Уравнение характеристик, полученное ранее:
$$
\displaystyle \frac{dt}{1}=- \frac{dx}{\frac{e^u}{1+e^{2u}}}=\frac{du}{0}
$$
Решая уравнение, получаем:
$$
\begin{cases}
   u = u_0 = const \\
   \displaystyle  \int\limits_{x_0}^{x}dx = - \int\limits_{t_0}^{t} \frac{e^u}{1+e^{2u}} dt
\end{cases}
$$
Интегрируем второе уравнение:
$$
\displaystyle x - x_{0} = - \frac{e^u}{1+e^{2u}}(t - t_{0})
$$
1. Пусть $t_0 = 0$, $-1 \leq x_0 < 1$.
Тогда:
$$
u_0 = - \sin(\pi x_0) \; \Rightarrow \;  \\ t = - \frac{1+e^{-2 \sin(\pi x_{0})}}{e^{- \sin(\pi x_{0})}} (x - x_{0})
$$
2. Пусть $x_0=0$, $t_0 \geq 0$. Тогда:
$$
u_0 = e^{-t_0}-1 \; \Rightarrow \; \\ x = - \frac{e^{e^{-t_{0}}-1}}{1+e^{2(e^{-t_{0}}-1)}} (t - t_{0})
$$

Построим проекции характеристик. Голубым цветом отмечены характеристики, соотвествующие выбору $x_0=0$, синим цветом - соответствующие выбору $t_0=0$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def when_t0_0(x0): # выражение справа, когда t0 = 0
    return 1 + np.exp(2)*np.exp(-np.sin(np.pi * x0))/(np.exp(-np.sin(np.pi * x0)))

def when_x0_0(t0): # выражение справа, когда x0 = 0
    return 1 + np.exp(2)*np.exp(np.exp(-t0) - 1)/(np.exp(np.exp(-t0) - 1))

x0, t0 = 0, 0 # пределы изменения x, t
x1, t1 = -1, 1
n, m = 16, 8 # в конечном счете - количество характеристик

x, t = np.linspace(x1, x0, n), np.linspace(t0, t1, m)  # задаем сетку


fig = plt.figure(figsize=(10,10))
fig, ax =plt.subplots(figsize=(10,10))
plt.title('Проекции характеристик на плоскость (x,t)', fontsize=11)
plt.xlim(x1, x0)
plt.ylim(t0,t1)
plt.xlabel('x', fontsize=11)
plt.ylabel('t', fontsize=11)


for x0 in x:
    for t0 in t:
        plt.plot(x, -(x - x0)*when_t0_0(x0), color = "blue", linewidth=1)
        plt.plot(x, t0 - x*when_x0_0(t0), color = "dodgerblue", linewidth=1)

#plt.plot(x, -8*x, color = "g") 
ax.tick_params(axis = 'both', which = 'major', direction ='in')

plt.show()
```

Видно, что в заданной области характеристики не пересекаются. Следовательно, в области $x \in [-1, 0)$, $t \in [0, 1]$ решение определено однозначно. При $t_0 = 0$ решение определяется начальными условиями, при $x_0=0$ - граничными условиями.

# Численное решение

Построим разностную схему. 

Введем в заданной области $\Omega = \{ (x,t):  -1 \leq x < 0, \; 0 \leq t \leq 1 \}$ разностную сетку:
$$\omega_{h \tau} = \{ x_{n} = nh, n = \overline{0,N} ; \; t_{m} = m \tau, m = \overline{0,M} \}$$
где $N$ - число узлов вдоль оси $x$, $M$ - число узлов вдоль оси $t$, $h$ - шаг по $x$, $\tau$ - шаг по $t$.

На сетке $\omega_{h \tau}$ будем рассматривать сеточную функцию $u^{m}_{n} = u(x_{n}, t_{m})$.

Выделим в исходном уравнении производную сложной функции:
$$
\displaystyle \frac{\partial u}{\partial t} - \frac{e^u}{1+e^{2u}} \frac{\partial u}{\partial x} = \frac{\partial u}{\partial t} - \frac{\partial(\arctan e^u)}{\partial x}
$$

Обозначим $F(u) = - \arctan e^u$, $F^m_n=F(u^m_n)$ - вспомогательная сеточная функция.

Для построения разностной схемы (схемы бегущего счета) будем использовать трехточечный шаблон, называемый нижним уголком:

<center><img src=https://i.postimg.cc/JhMPHfwq/n-ugol.png width="300"></center>

Тогда разностная схема для данной задачи будет иметь вид:
$$
\begin{cases}
   \displaystyle \frac{1}{\tau} \left(u^{m+1}_{n} - u^m_{n} \right) + \frac{1}{h} \left( F^m_n - F^m_{n-1} \right) = 0, \; \; n = \overline{1,N}, \; \; m = \overline{0,M-1} \\
   \displaystyle F^m_n = - \arctan e^{u^m_n} \\
   u^{0}_{n} = - \sin\pi x_{n}, \; \; n = \overline{0,N} \\
   u^m_{0} = e^{-t_{m}} - 1, \; \; m = \overline{0,M}
\end{cases}
$$

Задача решается при помощи схемы бегущего счета и итерационного метода Ньютона. Уравнение в неявном виде при $n=1$:
$$
\displaystyle f(u^{m+1}_1) = \frac{1}{\tau} \left(u^{m+1}_{1} - u^m_{1} \right) + \frac{1}{h} \left( F^m_1 - F^m_{0} \right) = 0 
$$

Пусть известно некоторое приближение $u^{(s)}_1$ к корню $u^{m+1}_1$. Тогда уравнение принимает вид:
$$
f(u^{(s)}_1 + \Delta u^{(s)}_1) = 0, \text{где} \; \Delta u^{(s)}_1 = u^{m+1}_1 - u^{(s)}_1  
$$

Разложим это уравнение в ряд Тейлора:
$$
f(u^{m+1}_1) = f(u^{(s)}_1 + \Delta u^{(s)}_1) = f(u^{(s)}_1) + f'(u^{(s)}_1) \Delta u^{(s)}_1 + o(u^{(s)}_1) = 0
$$
$$
\Delta u^{(s)}_1 + \frac{o(u^{(s)}_1)}{f'(u^{(s)}_1)} = - \frac{f(u^{(s)}_1)}{f'(u^{(s)}_1)} 
$$
$$
u^{m+1}_1 + \frac{o(u^{(s)}_1)}{f'(u^{(s)}_1)} - u^{(s)}_1 = u^{(s+1)}_1 - u^{(s)}_1 = \frac{f(u^{(s)}_1)}{f'(u^{(s)}_1)} 
$$
$$
u^{(s+1)}_1 = u^{(s)}_1 - \frac{f(u^{(s)}_1)}{f'(u^{(s)}_1)}
$$

Процесс останавливается, когда $|u^{(s+1)}_1 - u^{(s)}_1| < \varepsilon$, где $\varepsilon$ - заданная точность. Выбор $h$ и $\tau$ осуществляется с учетом устойчивости используемого метода. Для исследования устойчивости проведем дальнейшее рассмотрение.

# Порядок аппроксимации
Исследуем аппроксимацию исследуемой разностной схемы 
$$\psi = Lu - L_{h}u$$
где 
$$Lu = u_{t} + cu_{x} = u_{t} + F' u_{x} = u_t + F_x $$ 

$$\\ %\displaystyle L_{h}u = \frac{1}{\tau} \left( u(x_{n}, t_{m+1}) - u(x_{n}, t_{m}) \right) + \frac{c}{h} \left( u(x_{n}, t_{m}) - u(x_{n-1}, t_{m}) \right) \\ 
\displaystyle L_{h}u = \frac{1}{\tau} \left(u^{m+1}_{n} - u^m_{n} \right) + \frac{1}{h} \left( F^m_n - F^m_{n-1} \right) = \\
\displaystyle = \frac{1}{\tau} \left( u(x_{n}, t_{m+1}) - u(x_{n}, t_{m}) \right) + \frac{1}{h} \left( F(x_{n}, t_{m}) - F(x_{n-1}, t_{m}) \right)
$$
Здесь, как и ранее, $F=F(u) = - \arctan e^u, \; \; F'=F'(u)=- \frac{e^u}{1+e^{2u}}$ 

Вычислим $\psi$ во 2 узле, т.е. найдем
$$\psi|_{x_{n}, t_{m}} = Lu|_{x_{n}, t_{m}} - L_{h}u|_{x_{n}, t_{m}}$$
Для этого разложим функции в ряд Тейлора с центром в точке $(x_{n}, t_{m})$:
$$
u(x_{n}, t_{m+1}) = u(x_n,t_m) + u_t(x_n, t_m)\tau + u_{tt}(x_n,t_m)\frac{\tau ^2}{2} \\
F(x_{n-1}, t_{m}) = F(x_n,t_m) - F_x(x_n, t_m)h + F_{xx}(x_n,t_m)\frac{h^2}{2} \\
$$
Подставляя разложения в выражение для $L_h u|_{x_{n}, t_{m}}$, получаем:
$$
L_h u|_{x_{n}, t_{m}} = u_t(x_{n}, t_{m}) + u_{tt}(x_n,t_m) \frac{\tau}{2} + F_x(x_{n},t_{m}) - F_{xx}(x_n,t_m) \frac{h}{2} + O(\tau ^2) 
$$
Выражение для $Lu|_{x_{n}, t_{m}}$:
$$
Lu|_{x_{n}, t_{m}} = u_t(x_n,t_m) + F_x(x_n, t_m)
$$
Окончательно имеем:
$$
\psi|_{x_{n}, t_{m}} = u_t(x_n,t_m) + F_x(x_n, t_m) - u_t(x_{n}, t_{m}) - u_{tt}(x_n,t_m) \frac{\tau}{2} - F_x(x_{n},t_{m}) + F_{xx}(x_n,t_m) \frac{h}{2} + O(\tau ^2) 
$$
$$
= - \frac{\tau}{2} u_{tt}(x_n,t_m) + \frac{h}{2} F_{xx}(x_n,t_m) + O(\tau ^2 + h^2) \\ 
\Rightarrow \psi|_{x_{n}, t_{m}} = O(\tau + h)
$$
Следовательно, оператор $L_h$ аппроксимирует дифференциальный оператор $L$, порядок аппроксимации равен $1$.

# Исследование на устойчивость 
Для того, чтобы определить, является ли разностная схема устойчивой, воспользуемся необходимым и достаточным критерием устойчивости (условием Куранта).

Внесем погрешность в входные данные, т.е. пусть 
$$
f \rightarrow f + \delta f \Rightarrow u \rightarrow u + \delta u, \; \; \phi \rightarrow \phi + \delta \phi
$$
Тогда задача сведется к задаче относительно $\delta u$:
$$
\begin{cases}
   L \delta u = 0 \\
   l \delta u = \delta \phi
\end{cases}
$$
Выберем произвольную точку $(x^*, t^*) \in \omega_{h \tau}$ и пусть 
$$
c(x^*,t^*) = - \frac{e^{u(x^*,t^*)}}{1+e^{2u(x^*,t^*)}} = C
$$
Перейдем к разностной схеме:
$$
\begin{cases}
   \displaystyle \frac{1}{\tau} \left(\delta u^{m+1}_{n} - \delta u^m_{n} \right) + \frac{C}{h} \left( \delta u^m_n - \delta u^m_{n-1} \right) = 0, \\
   \delta u^{0}_{n} = - \delta (\sin\pi x_{n}) \\
   \delta u^m_{0} = 0
\end{cases}
$$
Разложим ошибку в ряд Фурье вида $\delta u^m_n = \lambda e^{i\omega n}$, где $\omega=qh$, $q \in \mathbb{R}$, а $\lambda$ - множитель роста. Подставляя в первое уравнение разностной схемы относительно $\delta u$, получаем:
$$
\displaystyle \frac{\lambda^{k+1} e^{i\omega n} - \lambda^{k} e^{i\omega n}}{\tau} + C \frac{\lambda^{k} e^{i\omega n} - \lambda^{k} e^{i\omega (n-1)}}{h}  = 0 \\
\displaystyle \frac{\lambda - 1}{\tau} + C \frac{\lambda - e^{-i\omega}}{h}  = 0 \\
\Rightarrow \lambda = \frac{1 + \frac{C\tau}{h} e^{-i\omega}}{1 + \frac{C\tau}{h}}
$$
Вычислим $|\lambda|^2$:
$$
\displaystyle |\lambda|^2 = \frac{(1 + \frac{C\tau}{h} \cos\omega)^2 + (\frac{C\tau}{h} \sin\omega)^2}{(1 + \frac{C\tau}{h})^2} = \frac{h^2 + 2hC\tau \cos\omega + C^2 \tau^2}{h^2 + 2hC\tau + C^2 \tau^2} \leq 1  \Rightarrow \\ |\lambda| \leq 1
$$
Как видно, необходимый и достаточный критерий устойчивости выполнен, следовательно схема является безусловно устойчивой.


# Сходимость решения
Как было показано ранее, разностная схема аппроксимирует задачу с порядком $1$. Более того, разностная схема является устойчивой. Применим следующую теорему:

Теорема. Пусть разностная схема линейна. Если она устойчива и аппроксимирует исходную задачу с некоторым порядком p, то она сходится, причем порядок сходимости равен p.

Следовательно, найденное численное решение исходной задачи сходится к аналитическому.

# Построение графиков

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Начальные условия по x
def N_U_t(x):
    return -np.sin(np.pi * x)

# Граничные условия по t
def G_U_x(t):
    return np.exp(-t) - 1

# F и F'
def F(u): 
    return -np.arctan(np.exp(u))

def deriv_F(u): 
    return -(np.exp(u)/(1 + np.exp(2*u))) 

# F и F' для итерационного метода. Обозначения:
# u1_n_1 = u(n-1, m+1), u1_n = u(n, m+1) 
# u_n_1 = u(n-1, m); u_n = u(n, m) 
def iter_F_box(u1_n, u_n, u1_n_1, u_n_1): 
    return (u1_n - u_n + u1_n_1 - u_n_1)/(2*τ) + (F(u_n) - F(u_n_1) + F(u1_n) - F(u1_n_1))/(2*h)

def iter_deriv_F_box(u): 
    return 1/(2*τ) + deriv_F(u)/(2*h)

def iter_F_n_ugol(u1_n, u_n, u_n_1):
    return (u1_n - u_n)/τ + (F(u_n) - F(u_n_1))/h

def iter_deriv_F_n_ugol(u):
    return 1/τ + deriv_F(u)/h

def iter_F_verh_ugol(u1_n, u_n, u1_n_1):
    return (u1_n - u_n)/τ + (F(u1_n) - F(u1_n_1))/h

def iter_deriv_F_verh_ugol(u):
    return 1/τ + deriv_F(u)/h

# Итерационный метод Ньютона (3 схемы)
def Newton_n_ugol(u_n, u_n_1):
    u1_n = u_n 
    d = ε + 1
    while (d > ε):
        X_T = u1_n
        u1_n = X_T - iter_F_n_ugol(u1_n, u_n, u_n_1)/iter_deriv_F_n_ugol(u1_n)
        d = abs(u1_n - X_T)
    return u1_n

def Newton_verh_ugol(u_n, u1_n_1):
    u1_n = u1_n_1
    d = ε + 1
    while (d > ε):
        X_T = u1_n
        u1_n = X_T - iter_F_verh_ugol(u1_n, u_n, u1_n_1)/iter_deriv_F_verh_ugol(u1_n)
        d = abs(u1_n - X_T)
    return u1_n

def Newton_box(u_n, u1_n_1, u_n_1):
    u1_n = u_n_1
    d = ε + 1
    while (d > ε):
        X_T = u1_n
        u1_n = X_T - iter_F_box(u1_n, u_n, u1_n_1, u_n_1)/iter_deriv_F_box(u1_n)
        d = abs(u1_n - X_T)
    return u1_n       
```

```python
# "n_ugol" - нижний уголок, "verh_ugol" - верхний уголок, "box" - ящик
SCHEME = "n_ugol" 

#n, m = 100, 100  # размерность сетки

x0, t0 = 0, 0 # пределы изменения x, t
x1, t1 = -1, 1

#значения n, m, для которых будут строиться графики
N_M = [(10, 10), (20, 20), (40, 40), (100, 100)]

fig = plt.figure(figsize=(20,40))
num = 1
#цикл для того, чтобы построить графики при разных n, m
for _just_for_loop_, step in enumerate(N_M):
    n, m = step
    
    X_T = np.zeros((n, m), dtype=float) # заполнение X_T
    x, t = np.linspace(x0, x1, n), np.linspace(t0, t1, m) 
    # равномерное заполнение от x0/t0 до x1/t1 n-мя/m-мя элементами

    h = float(x1 - x0)/(n-1) # шаг сетки по x
    τ = float(t1 - t0)/(m-1) # шаг сетки по t
    ε = 0.00001 #точность

    X_T[:, 0] = N_U_t(x)  # начальные условия по t
    X_T[0, :] = G_U_x(t)  # граничные условия по x

    
    for i in range(1, n):
        for j in range(1, m):
            if (SCHEME == "n_ugol") :
                X_T[i, j] = Newton_n_ugol(X_T[i, j-1], X_T[i-1, j-1])
            if (SCHEME == "verh_ugol") :
                X_T[i, j] = Newton_verh_ugol(X_T[i, j-1], X_T[i-1, j])
            if (SCHEME == "box") :
                X_T[i, j] = Newton_box(X_T[i, j-1], X_T[i-1, j], X_T[i-1, j-1])        
                
    t, x = np.meshgrid(t, x)
    ax = fig.add_subplot(4, 2, num, projection='3d')
    surf=ax.plot_surface(x, t, X_T, rstride=1, cstride=1, cmap = 'winter')
    plt.title('Численное решение при (n, m) = ' + str(step), fontsize=15)
    plt.ylabel('t', fontsize=15)
    plt.xlabel('x', fontsize=15)
    num += 1
    ax = fig.add_subplot(4, 2, num, projection='3d')
    surf=ax.plot_surface(t, x, X_T, rstride=1, cstride=1, cmap = 'winter')
    plt.title('Численное решение при (n, m) = ' + str(step), fontsize=15) 
    plt.ylabel('x', fontsize=15)
    plt.xlabel('t', fontsize=15)
    num += 1

plt.show()
```

# Сходимость численного решения 
Для оценки сходимости решения построим графики зависимости $u(x)$ при разных значениях $n$, $m$.


```python
x0, t0 = 0, 0 # пределы изменения x, t
x1, t1 = -1, 1

#значения n, m, для которых будут строиться графики
N_M = [(10, 10), (20, 20), (40, 40), (100, 100)]

fig_ = plt.figure(figsize=(20, 20))
num = 1

#цикл для того, чтобы построить графики при разных n, m
for _just_for_loop_, step in enumerate(N_M):
    n, m = step
    
    X_T = np.zeros((n, m), dtype=float) # заполнение X_T
    x, t = np.linspace(x0, x1, n), np.linspace(t0, t1, m) 
    # равномерное заполнение от x0/t0 до x1/t1 n-мя/m-мя элементами

    h = float(x1 - x0)/(n-1) # шаг сетки по x
    τ = float(t1 - t0)/(m-1) # шаг сетки по t
    ε = 0.00001 #точность

    X_T[:, 0] = N_U_t(x)  # начальные условия по t
    X_T[0, :] = G_U_x(t)  # граничные условия по x
    
    for i in range(1, n):
        for j in range(1, m):
            if (SCHEME == "n_ugol") :
                X_T[i, j] = Newton_n_ugol(X_T[i, j-1], X_T[i-1, j-1])
            if (SCHEME == "verh_ugol") :
                X_T[i, j] = Newton_verh_ugol(X_T[i, j-1], X_T[i-1, j])
            if (SCHEME == "box") :
                X_T[i, j] = Newton_box(X_T[i, j-1], X_T[i-1, j], X_T[i-1, j-1])        
                
    t, x = np.meshgrid(t, x)
    ax = fig_.add_subplot(2, 2, num)
    ax.plot(x, X_T[:,int(m/2)], color = "blue", linewidth=1)
    plt.title('График завимисти u(x), t = 0, при (n, m) = ' + str(step), fontsize=15)
    plt.ylabel('u', fontsize=15)
    plt.xlabel('x', fontsize=15)
    num += 1

plt.show()
```

Видно, что при сгущении сетки решение стремится к одной кривой, что  указывает на его сходимость.
